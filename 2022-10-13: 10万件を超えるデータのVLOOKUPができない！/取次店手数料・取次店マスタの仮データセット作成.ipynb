# 取次店手数料仮データセット生成

*   2021-10-01から2022-10-12の日付まで
*   自分のGoogle Driveに　「取次店手数料_2021年10月から2022年10月まで.xlsx」　ファイルを新しく作成する
*   中身は毎回ランダム



```python
import pandas as pd
import random
```

```python
dt_list_20211001_20221012 = pd.date_range(start='2021-10-01', end='2022-10-12', freq='D')
```

```python
dt_list_20211001_20221012
```

```python
dateList = list(dt_list_20211001_20221012)
```

```python
records = []
```

```python
for i in dateList:
    repeat = random.randrange(340, 1000)
    for j in range(repeat):
        date = i.strftime("%Y-%m-%d")
        customerId = random.randrange(10001, 20000)
        agencyId = random.randrange(100, 180)
        agencySubId = random.randrange(1, 8)
        agencyFeeAmount = random.randrange(-10, 35, 5) * random.randrange(1000, 10000, 1000)
        closing = random.randrange(0, 10)
        record = {
            '年月日': date
            , 'お客様番号': customerId
            , '取次店番号': agencyId
            , '取次店番号（サブ）': agencySubId
            , '取次店手数料（円）': agencyFeeAmount
            , '清算': '済み' if closing < 9 else '未清算'
        }
        records.append(record)
```

```python
len(records)
```

```python
df = pd.DataFrame(records)
```

```python
df.to_excel('drive/MyDrive/取次店手数料_2021年10月から2022年10月まで.xlsx', index=False)
```

# 取次店マスタ仮データセット生成

*   作られた　「取次店手数料_2021年10月から2022年10月まで.xlsx」　から取次店番号を取得
*   ユニークな番号ごとに仮のランダムな会社名などを付け加える
*   その結果を自分のGoogle Driveに　「取次店マスタ_202210.xlsx」　ファイルとして新しく作成する



```python
!pip install faker
```

```python
from faker import Faker
```

```python
datagenerator = Faker(['ja_JP'])
```

```python
agencyMaster = df[['取次店番号', '取次店番号（サブ）']].drop_duplicates()
```

```python
agencyMaster['取次店名'] = [datagenerator.company() for x in range(len(agencyMaster))]
```

```python
agencyMaster['支店名'] = agencyMaster['取次店番号（サブ）'].apply(lambda x: '本店' if x == 1 else datagenerator.city()[:-1] + '支店')
```

```python
agencyMaster['住所'] = [datagenerator.address() for x in range(len(agencyMaster))]
```

```python
agencyMaster['担当者'] = [datagenerator.name() for x in range(len(agencyMaster))]
```

```python
agencyMaster['会社メール'] = [datagenerator.company_email() for x in range(len(agencyMaster))]
```

```python
agencyMaster.to_excel('drive/MyDrive/取次店マスタ_202210.xlsx', index=False)
```
