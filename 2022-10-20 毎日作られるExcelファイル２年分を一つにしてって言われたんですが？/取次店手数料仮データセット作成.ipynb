```python
dt_list_20211001_20221019 = pd.date_range(start='2020-10-01', end='2022-10-19', freq='D')
```

```python
dateList = list(dt_list_20211001_20221019)
```

```python
records = []
```

```python
column1 = ['年月日', '年月日 ', '日付', 'Date']
column2 = ['お客様番号', 'お客様番号 ', '顧客番号', '顧客ID', 'お客さま番号', 'CustID']
column3 = ['取次店番号', '取次店番号 ', '商社番号', '代理店番号', '取次店', '取次店ID']
column4 = ['取次店番号（サブ）', '取次店番号（サブ） ', '支店', '支店番号', '取次店番号(Sub)', '取次店サブID', '取次店SubID']
column5 = ['取次店手数料（円）', '取次店手数料（円）  ', '手数料（円）', '手数料(円)', '手数料', 'JPY', 'AgencyFee']
column6 = ['清算', '精算', '精算　', '精算有無', 'Status']
```

```python
%time

for i in dateList:
    repeat = random.randrange(340, 1000)
    for j in range(repeat):
        date = i.strftime("%Y-%m-%d")
        customerId = random.randrange(10001, 20000)
        agencyId = random.randrange(100, 180)
        agencySubId = random.randrange(1, 8)
        agencyFeeAmount = random.randrange(-10, 35, 5) * random.randrange(1000, 10000, 1000)
        closing = random.randrange(0, 10)
        record = {
            column1[0]: date
            , column2[0]: random.choice([int(customerId), str(customerId)])
            , column3[0]: random.choice([int(agencyId), str(agencyId)])
            , column4[0]: random.choice([int(agencySubId), str(agencySubId)])
            , column5[0]: random.choice([int(agencyFeeAmount), str(agencyFeeAmount)])
            , column6[0]: random.choice(['済み', '済']) if closing < 9 else random.choice(['未清算', '未精算'])
        }
        records.append(record)
```

```python
len(records)
```

```python
df = pd.DataFrame(records)
```

```python
df
```

```python
%time

for date in dt_list_20211001_20221019:
    df_tmp = df[ df[column1[0]] == date.strftime("%Y-%m-%d") ]      \
        .rename(
            columns={
                column1[0]: random.choice(column1) 
                , column2[0]: random.choice(column2) 
                , column3[0]: random.choice(column3) 
                , column4[0]: random.choice(column4) 
                , column5[0]: random.choice(column5) 
                , column6[0]: random.choice(column6) 
            })                                                      \
        .to_excel(f'drive/MyDrive/20221020_QCELLS_Handson/{date.year}年{date.month:02d}月/{date.year}{date.month:02d}{date.day:02d}.xlsx', index=False)
```
