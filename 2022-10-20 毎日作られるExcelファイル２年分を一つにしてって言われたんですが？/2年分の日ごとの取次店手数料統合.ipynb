# Step3: 各ファイルの行数、カラム数、カラム名を確認します。
 - カラム名の空白が入っていたりカラム数があってないなど、何を直して合体させるべきかイメージをつかむためです。

```python
columnsMap = []
countTotalRows = 0
countColumnsEach = []
```

```python
for filePath in fileList:
  df_tmp = pandas.read_excel(filePath)
  print(len(df_tmp), df_tmp.columns)
  columnsMap.append(list(df_tmp.columns))
  countTotalRows = countTotalRows + len(df_tmp)
  countColumnsEach.append(len(df_tmp.columns))

%time
```

```python
columnsMap
```

```python
countTotalRows
```

```python
set(countColumnsEach)
```

# Step4: カラムの種類がいろいろなので、統一させる準備をしましょう。

```python
for i in zip(*columnsMap):
  print(set(i))
```

```python
columnsMaster = ['年月日 ', 'お客様番号', '取次店番号', '取次店番号（サブ）', '取次店手数料（円）', '精算']
```

```python
df_union = pandas.DataFrame(columns=columnsMaster)
```

```python
df_union
```

# Step5: 用意した枠にファイルの内容を全部入れましょう！

```python
for filePath in fileList:
  df_tmp = pandas.read_excel(filePath)
  df_tmp.columns = columnsMaster
  df_union = df_union.append(df_tmp, ignore_index=True)

%time
```

# Step7: データが正しいか確認、修正し結果をExcelファイルに保存しましょう

```python
df_editing[ df_editing.duplicated(keep=False) ]
```
